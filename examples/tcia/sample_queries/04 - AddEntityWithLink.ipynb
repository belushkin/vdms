```python
import vdms

db = vdms.VDMS()
db.connect("localhost")

# Add a doctor and connect with patient

query = """
[
      {
      "FindEntity" : {
         "class" : "patient",
         "results" : {
            "list" : [ "bcr_patient_barc", "gender", "age_at_initial"], 
            "sort": "age_at_initial"
         },
         "constraints": { "age_at_initial": [">=", 85]}, 
         "_ref": 35
      }
   },
   {
      "AddEntity" : {
         "class" : "doctor", 
         "properties": { 
             "age": 17, 
             "name": "Sammy", 
             "lastname": "Smith", 
             "speciality": "braisnstuff"
             
         },
         "link": {
             "ref": 35, 
             "class": "doctor_patient"
         }
      }
   }
]   
"""

response, images = db.query(query)
print "VDMS Response: \n"
vdms.aux_print_json(str(response))
```

```python
query = """
[
   {
      "FindEntity" : {
         "class" : "doctor", 
         "constraints": {
             "name": ["==", "Sammy"]
         }, 
         "_ref": 24
      }
   }, 
   {
      "FindEntity" : {
         "class" : "patient", 
         "link": {
             "ref": 24
         }, 
         "results": {
             "list": ["age_at_initial", "gender"]
         }
      }
   }
]   
"""

response, images = db.query(query)
print "VDMS Response: \n"
vdms.aux_print_json(str(response))
```
